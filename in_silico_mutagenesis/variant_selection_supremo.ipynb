```python
def determine_category(row):
    if row['Tumor_Seq_Allele1'] == '-' and set(row['Tumor_Seq_Allele2']).intersection({'A', 'C', 'T', 'G'}):
        return 'Insertion'
    elif row['Tumor_Seq_Allele2'] == '-' and set(row['Tumor_Seq_Allele1']).intersection({'A', 'C', 'T', 'G'}):
        return 'Deletion'
    else:
        length_diff = len(row['Tumor_Seq_Allele1']) - len(row['Tumor_Seq_Allele2'])
        if length_diff == 0:
            return 'Substitution'
        elif length_diff > 0:
            return 'Deletion'
        elif length_diff < 0:
            return 'Insertion'
        else:
            return 'Miscellaneous'

# Apply the function to create the 'category' column
variants_to_test['category'] = variants_to_test.apply(determine_category, axis=1)
```

```python
supremo_variants = supremo_variants.drop_duplicates(subset=['Hugo_Symbol', 'Chromosome', 
                                                            'Start_Position', 'End_Position'])

supremo_variants = supremo_variants[['Chromosome', 'Start_Position', 'End_Position', 'Tumor_Seq_Allele1', 'Tumor_Seq_Allele2', 'category', 'Hugo_Symbol']]
```

```python
supremo_variants['Tumor_Seq_Allele2'][supremo_variants['Tumor_Seq_Allele2'] == '-'] = '<DEL>'
```

```python
supremo_variants = supremo_variants[supremo_variants['category'] != 'Insertion']
supremo_variants['SVTYPE'] = ''
supremo_variants['SVTYPE'][supremo_variants['category'] == 'Deletion'] = 'DEL'

supremo_variants['SVLEN'] = ''
supremo_variants['SVLEN'][supremo_variants['category'] == 'Deletion'] = -1

supremo_variants = supremo_variants.drop(columns='category')

supremo_variants = supremo_variants[['Chromosome', 'Start_Position', 'Tumor_Seq_Allele1', 'Tumor_Seq_Allele2', 'End_Position', 'SVTYPE', 'SVLEN', 'Hugo_Symbol']]

supremo_variants.columns = ['CHROM', 'POS', 'REF', 'ALT', 'END', 'SVTYPE', 'SVLEN', 'Gene_Name']
```

```python
seq_length = 49152

gene_TSS_file = pd.read_csv('/pollard/home/aravi1/Enformer_genomic_regions_TSSCenteredGenes_FixedOverlapRemoval.csv')
gene_TSS_file = gene_TSS_file[['seqnames', 'gene_start', 'gene_name']]

gene_TSS_file = gene_TSS_file[gene_TSS_file['gene_name'].isin(supremo_variants['Gene_Name'])]

gene_TSS_file['starts'] = gene_TSS_file['gene_start'].astype(int) - (seq_length / 2)
gene_TSS_file['ends'] = gene_TSS_file['gene_start'].astype(int) + (seq_length / 2)
```

```python
supremo_variants = supremo_variants.rename(columns={'Gene_Name': 'gene_name'})

supremo_variants_genes = pd.merge(supremo_variants, gene_TSS_file, on='gene_name', how='left')
supremo_variants_genes = supremo_variants_genes.dropna()

supremo_variants_genes['difference'] = supremo_variants_genes['POS'].astype(int) - supremo_variants_genes['gene_start'].astype(int)
```

```python
supremo_variants_genes = supremo_variants_genes[abs(supremo_variants_genes['difference']) <= seq_length / 2]
np.random.seed(150)

supremo_variants_genes_test = supremo_variants_genes.sample(n=5000, )
```

```python
supremo_variants_genes_test = supremo_variants_genes_test.reset_index(drop=True)
supremo_variants_genes_test
```

```python
# Condition

supremo_variants_genes = supremo_variants_genes_test
condition = (supremo_variants_genes['SVTYPE'] == 'DEL')

# Update rows where SVTYPE is DEL
supremo_variants_genes.loc[condition, 'SVLEN'] = 1
supremo_variants_genes.loc[condition, 'REF'] = '-'
supremo_variants_genes.loc[condition, 'ALT'] = '-'
supremo_variants_genes.loc[condition, 'END'] = supremo_variants_genes.loc[condition, 'END'] + 1
```

```python
supremo_variants_genes['difference'] = supremo_variants_genes['difference'] * -1
```

```python
sequence_filenames = [f'sequence_' + str(i) for i in range(0, supremo_variants_genes.shape[0], 1)]
supremo_variants_genes = supremo_variants_genes.reset_index(drop=True)

for i in range(0, supremo_variants_genes.shape[0], 1):
    file = pd.DataFrame(supremo_variants_genes[['CHROM', 'POS', 'REF', 
                                                'ALT', 'END', 'SVTYPE', 'SVLEN']].iloc[i,:]).T
    
    name = '/pollard/home/aravi1/SuPreMo/CRC_variant_sequences/input_txt/' + sequence_filenames[i] + '.txt'
    file.columns = ['CHROM', 'POS', 'REF', 'ALT', 'END', 'SVTYPE', 'SVLEN']
    
    file.to_csv(name, sep='\t', index=None)


variant_TSS_diff = pd.concat([pd.Series(sequence_filenames), supremo_variants_genes['difference']], axis=1)
variant_TSS_diff.to_csv('/pollard/home/aravi1/SuPreMo/CRC_variant_sequences/variant_TSS_difference.tsv', sep = '\t', index = None, header=None)
supremo_variants_genes.to_csv('/pollard/home/aravi1/SuPreMo/CRC_variant_sequences/variant_info_MAF.tsv', sep = '\t', index = None)
```
